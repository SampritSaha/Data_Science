```python
import ast
def convert(obj):
    l=[]
    for i in ast.literal_eval(obj):
        l.append(i['name'])
    return l


def convert_cast(obj):
    l=[]
    counter = 0
    for i in ast.literal_eval(obj):
        if counter != 5:
            l.append(i['name'])
            counter = counter + 1
        else:
            break
    return l


def convert_crew(obj):
    l=[]
    for i in ast.literal_eval(obj):
        if i['job']=='Director':
            l.append(i['name'])
            break
    return l
```

```python
movies['genres']=movies['genres'].apply(convert)
movies.head()
```

```python
movies['keywords']= movies['keywords'].apply(convert)
movies.head()
```

```python
movies['cast']= movies['cast'].apply(convert_cast)
movies.head()
```

```python
movies['crew']= movies['crew'].apply(convert_crew)
movies.head()
```
